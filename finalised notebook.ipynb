# 1. Business Understanding

## 1.1 Business Overview

 >Telecommunication companies face increasing challenges in retaining customers due to market competition and changing consumer preferences. By analyzing customer data, such as demographics, service usage, contract type, and payment methods, these companies aim to identify patterns associated with churn. Understanding these patterns allows telecom providers to develop targeted strategies to reduce churn rates, enhance customer satisfaction, and optimize service offerings, ultimately driving sustainable growth and maintaining a competitive advantage in the dynamic telecom industry.

## 1.2 Problem Statement

 >Telecommunications companies face substantial revenue loss from high customer churn rates. Without understanding the factors driving churn through customer behavior and service data, they struggle to implement effective retention strategies, leading to higher acquisition costs and reduced profitability.

**Other Problems**

- Customer churn in the telecom industry is higher than average for many companies.
- Losing customers translates directly to lost revenue, impacting the bottom line.
- High churn rates can damage a company's reputation and make it harder to attract new customers
- Replacing lost customers is expensive, requiring more marketing and outreach efforts

## 1.3 Proposed Solutions

- Develop predictive models to identify customers at risk of churning based on behavior patterns and usage data, allowing for proactive retention efforts.
- Improve service quality and customer support by addressing common pain points, providing faster resolution of issues, and enhancing overall service delivery to increase customer satisfaction and loyalty.
- Understanding individual customer needs and tailoring services accordingly increases satisfaction.


## 1.4 Objectives

### 1.4.1 Main Objectives

>To identify predictable patterns in customer behavior and service usage that indicate a higher likelihood of churn, enabling Syriatel to implement targeted strategies for customer retention and reduce revenue loss from customer attrition.

### 1.4.2 Specific Objectives

 - Identify and segment customers who are most likely to churn based on their usage patterns, service history, and engagement levels
 - Determine the key drivers of customer dissatisfaction by analyzing customer service interactions, complaints, and feedback, and develop initiatives to improve overall service quality and satisfaction.
 - come up with strategies that will reduce customer's service calls by addressing common issues affecting their customers
 - Develop machine learning models that will predict wether a customer will churn or not
 - Aim to keep existing customers satisfied and loyal to the company's services.


## 1.5 Success Metrics

>Accuracy â€“ The proportion of correctly predicting wether a  customer will churn or not 75%
> ROC- AUC curve 80%




## 1.6 StakeHolders

- **Customers**: The primary stakeholders whose behavior and satisfaction directly impact churn rates.
- **Marketing Team**: Responsible for developing and implementing strategies to attract and retain customers
- **Finance Team**: Monitors the financial impact of churn on revenue and profitability
- **Customer Service Team**: Handles customer interactions and support
- **Employees** :They interact directly with customers and play a crucial role in delivering positive experiences.


## 1.6 Constraints

- Choosing the Right Metrics: Different stakeholders may prefer different evaluation metrics (e.g., accuracy, precision, recall, F1-score, AUC-ROC). Balancing these can be tricky, especially with imbalanced data.
- The churn column has class imbalance that might affect the modelling process where the models learn so well on the majority class than the minority class

### Appplying SMOTE and Hyperparameter Tuning

```python
from imblearn.over_sampling import SMOTE
X_train_resampled, y_train_resampled = SMOTE().fit_resample(X_train, y_train)
```

```python
# Define a function to train a model using GridSearchCV with cross-validation
def train_evaluate_model(model, params, X_train_resampled, y_train_resampled, X_test):
    # Step 1: Initialize GridSearchCV with the specified model and parameters
    grid_search = GridSearchCV(model, params, cv=5, scoring='roc_auc')

    # Step 2: Train the model on the resampled data
    grid_search.fit(X_train_resampled, y_train_resampled)

    # Step 3: Get the best model from GridSearchCV
    best_model = grid_search.best_estimator_

    # Step 4: Predict using the best model on the test set
    y_pred = best_model.predict(X_test)

    return best_model, y_pred
```

#### Logistic Regression

```python
# Define hyperparameters for LogisticRegression
logistic_params = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000],
                   'penalty': ['l2'],
                   'solver': ['liblinear'],
                   'class_weight': ['balanced']}
# Train and evaluate the Logistic Regression model
logistic_model, logistic_pred = train_evaluate_model(LogisticRegression(), logistic_params, X_train_resampled, y_train_resampled, X_test)
```

#### Random Forest

```python
# Random Forest
rf_params = {'n_estimators': [100, 200, 300], 'max_depth': [None, 10, 20, 30]}
rf_model, rf_pred = train_evaluate_model(RandomForestClassifier(), rf_params, X_train_resampled, y_train_resampled, X_test)
```

```python
# Gradient Boosting
gb_params = {'n_estimators': [100, 200, 300], 'learning_rate': [0.01, 0.1, 0.2]}
gb_model, gb_pred = train_evaluate_model(GradientBoostingClassifier(), gb_params, X_train_resampled, y_train_resampled, X_test)
```

```python
# Function to train and evaluate a model using GridSearchCV with cross-validation
def train_evaluate_model(model, param_grid, X_train, y_train, X_test, y_test):
    # Step 1: Resample the data using SMOTE
    X_train_resampled, y_train_resampled = SMOTE().fit_resample(X_train, y_train)

    # Convert y_train_resampled to integers (if it's not already)
    y_train_resampled = y_train_resampled.astype(int)

    # Step 2: Initialize GridSearchCV with the specified model and parameters
    cv = StratifiedKFold(n_splits=5)  # Stratified K-Fold cross-validation
    grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=cv, scoring='roc_auc')

    # Step 3: Train the model on the resampled data
    grid_search.fit(X_train_resampled, y_train_resampled)

    # Step 4: Get the best model from GridSearchCV
    best_model = grid_search.best_estimator_

    # Step 5: Predict and evaluate the model on the test set
    y_pred = best_model.predict(X_test)
    y_pred_proba = best_model.predict_proba(X_test)[:, 1]  # Probabilities for ROC AUC

    # Print evaluation metrics
    print("Best Parameters:", grid_search.best_params_)
    print("Accuracy on Test Set:", accuracy_score(y_test, y_pred))
    print("Precision on Test Set:", precision_score(y_test, y_pred))
    print("Recall on Test Set:", recall_score(y_test, y_pred))
    print("F1 Score on Test Set:", f1_score(y_test, y_pred))
    print("ROC AUC Score:", roc_auc_score(y_test, y_pred_proba))
    print("Classification Report:\n", classification_report(y_test, y_pred))
    print("Train Accuracy:", accuracy_score(y_train_resampled, best_model.predict(X_train_resampled))) # Predict on resampled data for train accuracy
    # print(classification_report(y_test, y_pred))
    conf_matrix = confusion_matrix(y_test, y_pred)
    sns.heatmap(conf_matrix, annot=True, fmt='d')
    plt.title(f'{model} Confusion Matrix')
    plt.show()
    return best_model, y_pred
```
